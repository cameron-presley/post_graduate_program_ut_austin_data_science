# build a function for categorial variables to address repetitive operations

```python
#Cite Reference:  this snipet of code (function) is directly from the Great Learning UBER case study and reviewed with Shanky on 2021-10-23
#The goal of this code is to productionize repetitive code for any variable
# and imrove the efficiency of our coding + storage 
#this is an atlernative approach to all the individual bar chart via box plot commands
#I changed the figure size to better fit with other graphics

# Function to create barplots that indicate percentage for each category.
def bar_perc(plot, feature):
    '''
    plot
    feature: 1-d categorical feature array
    '''
    total = len(feature) # length of the column
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/total) # percentage of each class of the category
        x = p.get_x() + p.get_width() / 2 - 0.05 # width of the plot
        y = p.get_y() + p.get_height()           # hieght of the plot
        ax.annotate(percentage, (x, y), size = 14) # annotate the percantage



```

```python
#t_stat, pval =  stats.ttest_ind(sample1, sample2, equal_var=False)
#critical_value  = stats.t.ppf(1 - 0.05, df)
#If t_stat > critical_value Reject Null Hypothesis
#p = 1 - stats.t.cdf(t_stat, df)
#If p < alpha  # Reject Null Hypothesis


#need to extract the smoker sample set and the non-smoker sample set from the overall data set

smoker_sample=subj_dummies[["charges","modified_yes","modified_no"]]
smoker_sample.head()


```

```python
#extract only the smokers

smokers_only = smoker_sample.loc[smoker_sample.modified_yes > 0,:]
smokers_only.head()

```

```python
n1 = smokers_only["charges"].count()
print (n1)
```

```python
xbar1 = smokers_only["charges"].mean()
print (xbar1)
```

```python
std1 = smokers_only["charges"].std()
print (std1)
```

```python
#explore non-smokers

non_smokers = smoker_sample.loc[smoker_sample.modified_no > 0,:]
non_smokers.head()
```

```python
n2 = non_smokers["charges"].count()
print(n2)
```

```python
xbar2 = non_smokers["charges"].mean()
print (xbar2)
```

```python
std2 = non_smokers["charges"].std()
print (std2)
```

```python
#calculate the T - statistic for 2 samples with unqual variance

xabs = abs (xbar1-xbar2)
print (xabs)

S = std1 + std2
print (S)

denom = math.sqrt(S)
print (denom)


T= xabs/denom
print(T)
```
