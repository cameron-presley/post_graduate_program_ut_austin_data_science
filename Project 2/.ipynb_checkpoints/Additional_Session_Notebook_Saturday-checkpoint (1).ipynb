# Bayes Theorem

Bayes Theorem provides a principled way for calculating a conditional probability.

It is a deceptively simple calculation, although it can be used to easily calculate the conditional probability of events where intuition often fails.

Although it is a powerful tool in the field of probability, Bayes Theorem is also widely used in the field of machine learning. Including its use in a probability framework for fitting a model to a training dataset, referred to as maximum a posteriori or MAP for short, and in developing models for classification predictive modeling problems such as the Bayes Optimal Classifier and Naive Bayes

# Problem Statement

A local bank reviewed its credit card policy with the intention of recalling some of its credit cards. In the past approx. 5% of cardholders defaulted, leaving the bank unable to collect the outstanding balance. Hence, management established a prior probability of 0.05 that any particular cardholder will default. The bank also found that the probability of missing a monthly payment is .20 for customers who do not default. The probability of missing a monthly payment for those whose default is 0.5.

**Q) Given that a customer missed one or more monthly payments, compute the probability
that a customer will default**

**Bayes' Theorem**

</img>



P(Default|Missed) = P(Missed|Default) * P(Default) / (P(Missed|Default) * P(Default) + P(Missed|Not-Default) * P(Not-Default))

P(Missed|Not-Default) : Probability of missing the payment given that the customer doesn't default

```python
probab_default= 0.05
probab_missed_and_no_default=0.20
probab_missed_given_default=0.50
```

```python
probability_of_missed= (probab_missed_given_default*probab_default) + (probab_missed_and_no_default*(1-probab_default))
```

```python
probab_default_given_missed= (probab_missed_given_default * probab_default )/ probability_of_missed
```

```python
probab_default_given_missed
```

# Poisson Distribution

**A bank is interested in studying the number of people who use the ATM located outside during night hours. On an average, 1.6 customers walk up to the ATM during a 10-minute interval, between 9 pm and midnight.**

```python
n=3
rate=1.6 #lambda
```

**a) Find the probability of exactly 3 customers using the ATM in a 10-minute interval.**

```python
poisson = stats.poisson.pmf(n,rate)
print('Probability of exactly 3 customers using the ATM in a 10-minute interval is %1.2f' % poisson)
```

**b) What is the probability of 3 or fewer customers using the ATM?**

```python
n=np.arange(0,4)
rate=1.6
poisson = stats.poisson.pmf(n,rate)
#To calculatee the probability of 3 or fewer we will add the prob of 0,1,2 and 3 people
print("Probability of 3 or fewer customers using the ATM",poisson[0]+poisson[1]+poisson[2]+poisson[3])

#or

print("Probability of 3 fewer customers using the ATM",stats.poisson.cdf(3,1.6))
```

# Normal Distribution
**A radar unit is used to measure speeds of cars on a Highway. The speeds are normally distributed with a mean of 70 km/hr and a standard deviation of 10 km/hr.**

```python
xbar=70
sigma=10
```

**a) What is the probability that a car picked at random is traveling at more than 100 km/hr?**

```python
x=100
```

```python
z = (x - xbar)/sigma
print("part a:",1 - (stats.norm.cdf(z)))
```

**b) What is the probability that the car speed is between 80 Km / hr and 100 Km / hr**

```python
x=80
```

```python
z1 = (x-xbar)/sigma
print("part b:",stats.norm.cdf(z) - stats.norm.cdf(z1))
```

# Hypothesis Formulation

# Problem Statement

The mass of N1=20 acorns from oak trees up wind from a coal power plant and N2=30 acorns from oak trees down wind from the same coal power plant are measured. Are the acorns from trees down wind less massive then the ones from up wind? The sample sizes are not equal but we will assume that the population variance of sample 1 and sample 2 are equal.

sample up wind:
x1 = [10.8, 10.0, 8.2, 9.9, 11.6, 10.1, 11.3, 10.3, 10.7, 9.7, 7.8, 9.6, 9.7, 11.6, 10.3, 9.8, 12.3, 11.0, 10.4, 10.4]

sample down wind:
x2 = [7.8, 7.5, 9.5, 11.7, 8.1, 8.8, 8.8, 7.7, 9.7, 7.0, 9.0, 9.7, 11.3, 8.7, 8.8, 10.9, 10.3, 9.6, 8.4, 6.6, 7.2, 7.6, 11.5, 6.6, 8.6, 10.5, 8.4, 8.5, 10.2, 9.2]

## Step 1: Define null and alternative hypotheses

* H<sub>0</sub>: x&#772;<sub>1</sub> = x&#772;<sub>2</sub>, or x&#772;<sub>2</sub> - x&#772;<sub>1</sub> = 0, There is no difference between the means
* H<sub>A</sub>: x&#772;<sub>2</sub> < x&#772;<sub>1</sub>, or x&#772;<sub>2</sub> - x&#772;<sub>1</sub> < 0, There is difference between the means

## Step 2: Decide the significance level

α= 0.05

## Step 3: Identify the test

This is a two-sample, paired, t - test

## Step 4: Calculate the test-statistics and p-value

```python
# sample up wind
x1 = [10.8, 10.0, 8.2, 9.9, 11.6, 10.1, 11.3, 10.3, 10.7, 9.7, 
      7.8, 9.6, 9.7, 11.6, 10.3, 9.8, 12.3, 11.0, 10.4, 10.4]

# sample down wind
x2 = [7.8, 7.5, 9.5, 11.7, 8.1, 8.8, 8.8, 7.7, 9.7, 7.0, 
      9.0, 9.7, 11.3, 8.7, 8.8, 10.9, 10.3, 9.6, 8.4, 6.6,
      7.2, 7.6, 11.5, 6.6, 8.6, 10.5, 8.4, 8.5, 10.2, 9.2]


# NOTE: the p-value given is two-sided so the one-sided p value would be p/2

t, p_twotailed = stats.ttest_ind(x2, x1)
print("tstats = ",t, ", p_twotailed = ", p_twotailed, ", p_onetailed =", p_twotailed/2)
```

## Step 5:  Decide to reject or accept null hypothesis

Since P-value(0.003) < 0.05 (alpha) we reject the null hypothesis and conclude that there is a difference in the two means.

# Chi-Sq test

**Two companies A and B were mergerd. After the first appraisal cycle post merger, employees orignally belonging to Company B put an allegation that the management favors employees who were orignally a part of Company A. At 95% confidence, perform a hypothesis test to validate if the claim holds good.**

```python
df = pd.DataFrame({'Promoted': [15, 16], 'Not-promoted': [9, 15]}, index = ['Company A', 'Company B'])
```

```python
df
```

## Step 1: Define null and alternate hypotheses

H0: Promotions are dependent on Company 

H1: Promotions are independent of Company

## Step 2: Decide the significance level

Here we select α= 0.05 as per 95% Confidence Level requirement in the question.

## Step 3: Identify the test statistic

This is a Chi-sq Test where categorical data has been reported in raw frequencies

## Step 4: Calculate the p - value and test statistic

```python
chi2, pval, dof, exp_freq = chi2_contingency(df, correction = False)
```

```python
pval
```

## Step 5:  Decide to reject or fail to reject null hypothesis

Since the pvalue is > 0.05, therefore, at 95% confidence we fail to reject the null hypothesis which implies the management is not biased in favor of employees originally belonging to company A.

# Formulate the hypothesis of the One-way ANOVA of 'gender' variable with the 'weight6weeks' variable.

$H_0$ : The means of 'weight6weeks' variable with respect to each gender is equal.

$H_1$ : At least one of the means of 'weight6weeks' variable with respect to each gender is unequal.

```python
sns.boxplot(x="gender", y="weight6weeks", data=df)
plt.grid()
plt.show()
```

# One-way ANOVA with the 'gender' variable.

```python
formula = 'weight6weeks ~ C(gender)'
model = ols(formula, df).fit()
aov_table = anova_lm(model)
print(aov_table)
```

Since the p value in this scenario is less than $\alpha$ (0.05), we can say that we reject the Null Hypothesis ($H_0$).

# Formulate the hypothesis of the One-way ANOVA of 'Diet' variable with the 'weight6weeks' variable.

$H_0$ : The means of 'weight6weeks' variable with respect to each diet category is equal.

$H_1$ : At least one of the means of 'weight6weeks' variable with respect to each diet category is unequal.

```python
sns.boxplot(x="diet", y="weight6weeks", data=df)
plt.grid()
plt.show()
```

# One-way ANOVA with the 'Diet' variable.

```python
formula = 'weight6weeks ~ C(diet)'
model = ols(formula, df).fit()
aov_table = anova_lm(model)
print(aov_table)
```

Since the p value in this scenario is greater than $\alpha$ (0.05), we can say that we fail to reject the Null Hypothesis ($H_0$).

# Formulate the hypothesis of ANOVA with both 'gender' and 'Diet' variables with respect to the variable 'weight6weeks'.

$H_0$ : The means of 'weight6weeks' variable with respect to each diet category and gender is equal.

$H_1$ : At least one of the means of 'weight6weeks' variable with respect to each diet category and gender is unequal.

# ANOVA with both 'gender' and 'Diet' variables with respect to the variable 'weight6weeks'.

```python
formula = 'weight6weeks ~ C(gender) + C(diet)'
model = ols(formula, df).fit()
aov_table = anova_lm(model)
print(aov_table)
```

Considering both the factors(diet and gender), gender is a significant factor as the p value is <0.05 wheras diet is not a significant variable as p value of diet is >0.05
