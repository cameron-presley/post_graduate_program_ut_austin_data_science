# Build a function for categorial variables to address repetitive operations

```python
#Cite Reference:  this snipet of code (function) is directly from the Great Learning UBER case study and reviewed with Shanky on 2021-10-23
#The goal of this code is to productionize repetitive code for any variable
# and imrove the efficiency of our coding + storage 
#this is an atlernative approach to all the individual bar chart via box plot commands
#I changed the figure size to better fit with other graphics

# Function to create barplots that indicate percentage for each category.
def bar_perc(plot, feature):
    '''
    plot
    feature: 1-d categorical feature array
    '''
    total = len(feature) # length of the column
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/total) # percentage of each class of the category
        x = p.get_x() + p.get_width() / 2 - 0.05 # width of the plot
        y = p.get_y() + p.get_height()           # hieght of the plot
        ax.annotate(percentage, (x, y), size = 14) # annotate the percantage

```

# 2. Prove (disprove) that the medical claims made by people who smoke is greater than those who don't smoke. [HINT: Formulated a hypothesis and prove/disprove it]

Step 1. Look at Charges for Smokers and Non-Smokers



```python
#explore charges vs. smoker


plt.figure(figsize=(10,5))           
sns.boxplot(subj['smoker'],subj['charges'], palette ="crest")
plt.ylabel('charges')
plt.xlabel('smokers')
```

Observations:

1.  We see that there is a significant difference in the charges between those who Smoke and those who do not Smoke (i.e., non-Smokers).

2.  The Non-Smokers have a noticeable amount of outliers with respect to charges, however, the Grand Max barely reaches Q3 of the Smokers IQR for the sample in question. 

3. There are no outliers for the Smoker's Charges and the charges are right skewed.  

4.  The two variables are Mutually Exclusive.  In short, one cannot be a Smoker and be a Non-Smoker at the same time.  





# Step 1:  Define the Null and Alternative Hypothesis:


Null Hypothesis = there are is no difference between the charges incurred by Smokers and Non-Smokers

Alternate Hypothesis = there is a difference between the charges incurred by Smokers as compared to Non-Smokers.  Specifically, the charges incurred by Smokers is greater than those incurred by Non-Smokers.


Ho = xbar2 =xbar 1, or xbar2 - xbar1 = 0
Ha = xbar 2 > xbar1, or xbar2-xbar1 > 0





# Step 2:  Decide the significance level

Assume a 95% confidence level, or alpha = 0.05



# Step 3:  Identify the test statistic

Independent t test:  namely there are 2 samples that are related - smokers and no-smokers

Independent samples exist if no case or person from one group can be assigned to a case or person from the other group. The t-test for independent samples is also called the two-sample t-test or unpaired t-test.

In this case, Smokers and Non-Smokers cannot be in the same group.

If one of the samples had been measured at different times, the they would have been relative and we should use the Paired t-test.



# Step 4:  Calculate the p-value and test statistic, validate assumptions for suitability of test

```python
#separate the Smokers into a separate dataframe for analysis in the t-test
subj_smoker =subj.query("smoker== 'yes'")
subj_smoker.head()
```

```python
#separate the Non-Smokers into a separate dataframe for analysis in the t-test
subj_non_smoker = subj.query ("smoker=='no'")
subj_non_smoker.head()
```

```python
#Create an array of Charges for Smokers - need a one-dimensional data set to run Levene's test

smokers_charges = subj_smoker['charges'].to_list()

#Create an array of Charges for Non-Smokers

non_smokers_charges = subj_non_smoker['charges'].to_list()
```

# LEVENE'S Test for Equal Variances Assumption

Before we run the Test Statistic, we need to see if it is safe to assume that the samples have EQUAL Variance.  Need to run Levene's test.  Using the same significance level of 0.05, we will test the Null Hypothesis that the sample data have equal variances.


```python
#Levene's test centered at the median
stats.levene(smokers_charges,non_smokers_charges, center='median')

```

```python

#Levene's test centered at the mean
stats.levene(smokers_charges,non_smokers_charges, center='mean')
```

The p-values for Levene's test are essentially zero (i.e., < 0.05), therefore we must reject the Null Hyphothesis that the samples have an equal variance - or the assumption of homoscedasticity.

We now move to the Welch's t-test for samples with unequal variances.



# Welch's t-test

```python
# we are looking for a p-value for a greater than scenario, so it will be a one-side p-value, or p/2
# we use the same stats.ttest_ind, but set the flag for equal_var = False to indicate unequal variances
t, p_twotailed = stats.ttest_ind(smokers_charges, non_smokers_charges, equal_var = False)
```

```python
print ('tstats = ,',t)
print ('p_twotailed = ', p_twotailed)
rhs_p_value = p_twotailed/2
print ('rhs_p_value =', rhs_p_value)

```

# Conclusions:

The one-tailed p-value is much less than alpha, in this case, it is essentially ZERO. Therefore we will REJECT the Null Hypothesis.  In short we cannot conclude that there is "No Difference" in the Charges between Smokers and Non-Smokers.  Or in other words, we have a 95% confidence that the Charges associated with Smokers are different and in this case, greater than those associated with Non-Smokers.

```python
#explore bmi vs. sex

plt.figure(figsize=(10,5))           
sns.boxplot(subj['sex'],subj['bmi'], palette ="YlGn")
plt.ylabel('bmi')
plt.xlabel('sex')
```

# Step 1:  Define the Null and Alternate Hypothesis

Null Hypothesis - H0 = the BMI for Men and Women are equal = xbar1=xbar2 or xbar2-xbar1=0

Alternate Hypothesis - HA = the BMI for Men is greater than the BMI for Women, xbar2>xbar1 or xbar2-xbar1>0






# Step 2: Decide the significance level
Assume a 95% confidence level, or alpha = 0.05


# Step 3: Identify the test statistic
Apply 2-sample t-test:  independent - Men and Women



# Step 4:  Calculate the p-value and test statistic, validate assumptions for suitability of test

```python
#separate the Men and Women to explore BMI.  Create a Male only BMI df
subj_male =subj.query("sex == 'male'")
subj_male.head()

```

```python
#separate the Men and Women to explore BMI.  Create a Female only BMI df
subj_female =subj.query("sex == 'female'")
subj_female.head()
```

# LEVENE'S Test for Equal Variances Assumption
Before we run the Test Statistic, we need to see if it is safe to assume that the samples have EQUAL Variance. Need to run Levene's test.

```python
#Need a one-dimensional data set to run Levene's test
#Create an array of BMI for Males

male_bmi = subj_male['bmi'].to_list()


#Create an array of BMI for Females

female_bmi = subj_female['bmi'].to_list()
```

```python

#Levene's test centered at the mean
stats.levene(male_bmi,female_bmi, center='mean')
```

```python
#Levene's test centered at the median
stats.levene(male_bmi,female_bmi, center='median')

```

Using a significance level of 0.05, the p-value is greater than and we can accept the Null Hypothesis that the samples have equal variances.

```python
# we are looking for a p-value for a greater than scenario, so it will be a one-side p-value, or p/2
# we can assume equal variances

t_bmi, p_twotailed_bmi = stats.ttest_ind(male_bmi,female_bmi)
```

```python
print ('tstats = ,',t_bmi)
print ('p_twotailed = ', p_twotailed_bmi)
rhs_p_value_bmi = p_twotailed_bmi/2
print ('rhs_p_value =', rhs_p_value_bmi)
```

# Conclusion

* The p-value is < our significance level of 0.05, therefore we REJECT the Null Hypothesis that the mean value of BMI for Males and Females are the same.  Or, we can say with a 95% confidence level that the BMI for Males is greater than Females.

# 4.  Is the proportion of smokers significantly different across regions (different regions)?
[HINT: Create a cotingency table / cross tab, use the function: stats.chi2_contingency () ]

```python
#How many Smokers vs. Non-Smoker by region....

sns.countplot(subj['region'], hue=subj['smoker'], palette ='crest')
plt.show()

```

# Step 1:  Define the Null and Alternate Hypothesis

* H0:  The proportion of smokers is the equal across regions.
* HA:  The proportion of smokers across different regions is not equal (i.e., the proportion of smokers is significantly different across regions.)



# Step 2: Decide the significance level

* Apply a significance level of 0.05 to correspond with a 95% Confidence Level.



# Step 3:  Identify the Test Statistic
 
* We will apply the Chi-sq test where categorical data for smokers and regions is available in proportions.  

# Step 4:  Calculate the p-value and test statistic, validate assumptions for suitability of test

* We need to build a contingency table of proportions by smoker and region



```python
#build a contingency table for Chi-sq analysis :  use a table of frequencies and proportions for comparison
```
